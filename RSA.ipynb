```python
import random
import math

def exp_func(x, y,mod):
    exp = bin(y).replace("0b", "")  
    value = 1
    y=1
    for i in range(len(exp)-1,-1,-1):      
        if(exp[i]=='1'):
            value =(value*x)%mod
            x=value
            y=y*value
        else:
            value = (value * x)%mod
            x=value
    return y%mod

def get_coprime(n):
    while True:
        coprime = random.randrange(1,n)
        if math.gcd(coprime, n) == 1:
            return True
        else:
            return False

def fermat_primality(n):
    for i in range(1000):
        if (get_coprime(n)== False):
            return False
    return True

import pandas as pd
import math
data=[]
# extended Euclidean Algorithm
def gcdExtended(r1, r2, t1, t2):
    x=r1
    r=t=b=1
    print('q    ','r1     ','r2     ','r     ','t1     ','t2     ','t')
    while(r2>0):
        q=math.floor(r1/r2) 
        r=r1-q*r2
        t=t1-q*t2
        print(q,"    ",r1,"    ",r2,"    ",r,"    ",t1,"    ",t2,"    ",t)
        r1=r2
        r2=r           
        t1=t2
        t2=t
    if(r1==1):
        b=t1%x                
        return b
    else:
        print("Inverse not exist")
        return False
    
    

def generate_keypair(p, q):
    if not (fermat_primality(p) and fermat_primality(q)):
        raise ValueError('Both numbers must be prime.')
    elif p == q:
        raise ValueError('p and q cannot be equal')
    if(fermat_primality(p) and fermat_primality(q)):
        print(p ,"and",q,"numbers are both prime")
    n = p * q

    phi = (p-1) * (q-1)

    e = random.randrange(1, phi)
    t1 = 0
    t2 = 1
    d = gcdExtended(phi,e, t1, t2) 
    if(d==False):
        return False
    return ((e, n), (d, n))

def encrypt(pk, plaintext):
    key, n = pk
    cipher = exp_func(plaintext, key,n)      
    return cipher

def decrypt(pk, ciphertext):
    key, n = pk
    plain = exp_func(ciphertext, key,n)
    return plain

if __name__ == '__main__':
    print( "RSA Encrypter/ Decrypter")
    p = int(input("Enter a prime number (17, 19, 23, etc): "))
    q = int(input("Enter another prime number (Not one you entered above): "))
    
    print("\n")
    print ("Generating your public/private keypairs now . . .")
    public, private = generate_keypair(p, q)
    print("\n")
    print ("Your public key is ", public ," and your private key is ", private)
    message = int(input("Enter a message to encrypt with public key: "))
    encrypted_msg = encrypt(public, message)
    print ("Your encrypted message with public key: ",public,". . .")
    print("Encrypted message is",encrypted_msg)
    print("\n")
    print( "Decrypting message with private key ", private ," . . .")
    print ("Your message is:")
    print (decrypt(private, encrypted_msg))
```
