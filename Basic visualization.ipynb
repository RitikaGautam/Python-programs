```python
#1. Load the csv file of dataset into dataframe in Python program.
import pandas as pd
pf=pd.read_csv("KCLT.csv")
print(pf)
```

```python
#2. Find the mean, median, mode, min and max for all numeric attributes.
from statistics import mode
import pandas as pd
pf=pd.read_csv("KCLT.csv")
for column in pf:
    if column!='date':
        mean1=pf[column].mean()
        median1=pf[column].median()
        mode1=pf[column].mode()
        min1=pf[column].min()
        max1=pf[column].max()
        print("mean of column",column,"=",mean1)
        print("median of column",column,"=",median1)
        print("mode of column",column,"=",mode1)
        print("min of column",column,"=",min1)
        print("max of column",column,"=",max1)
        print("\n")
```

```python
#3. Plot	min and max for actual_mean_temp, actual_min_temp and 	
#actual_max_temp using bar graph. Use blue for min and red for max. 	
import numpy as np
import matplotlib.pyplot as plt

min1=pf["actual_mean_temp"].min()
min2=pf["actual_min_temp"].min()
min3=pf["actual_max_temp"].min()
max1=pf["actual_mean_temp"].max()
max2=pf["actual_min_temp"].max()
max3=pf["actual_max_temp"].max()
labels = ['actual_mean_temp','actual_min_temp','actual_max_temp']
# data to plot
n_groups = 3
minvalues = (min1,min2,min3)
maxvalues=(max1,max2,max3)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35

rects1 = plt.bar(index-bar_width/2, minvalues, bar_width,color='b',label='min')
rects2 = plt.bar(index + bar_width/2, maxvalues, bar_width,color='r',label='max')

plt.title('MIN MAX GRAPH')
ax.set_xticks(index)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()
plt.show()
```

```python
#4. Plot	two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015.
pf['date'] = pd.to_datetime(pf['date'])

amt_aug_14 = pf['actual_mean_temp'].loc[(pf['date'].dt.month == 8) & (pf['date'].dt.year == 2014)]
amt_mar_15 = pf['actual_mean_temp'].loc[(pf['date'].dt.month == 3) & (pf['date'].dt.year == 2015)]

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.hist(amt_aug_14, color='blue')
ax2.hist(amt_mar_15, color='yellow')

ax1.set(xlabel = 'actual_mean_temp')
ax2.set(xlabel = 'actual_mean_temp')
ax1.set_title("aug 2014")
ax2.set_title("mar 2015")
ax1.grid('true')
ax2.grid('true')
plt.show()
#key observations: histogram for aug 2014 shows that the actual_mean_temp during the month was mostly 77-78
#                  histogram for mar 2015 shows that the actual_mean_temp during the month was mostly 54-55

```

```python
#6. Find the correlation between the following pairs of attributes: 
#actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp.

from scipy.stats import pearsonr
corr1,_ = pearsonr(pf['actual_mean_temp'],pf['actual_min_temp'])
print('Coorelation between actual_mean_temp and actual_min_temp is %.3f' % corr1)
corr2,_ = pearsonr(pf['actual_mean_temp'],pf['actual_max_temp'])
print('Coorelation between actual_mean_temp and actual_max_temp is %.3f' % corr2)
#key observations: Correlation between both the group of variables is highly positive showing they are highly correlated
#                  (change in one led to same direction change in another variable)
#Pearson Correlation Coefficient = covariance(X,Y)/std(X)*std(Y)
#                                  , where Covariance(X,Y) = (sum (x - mean(X)) * (y - mean(Y)) ) * 1/(n-1)
# A value between -1 to 1 would give us a interpretable correlation
```
